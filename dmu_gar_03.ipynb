```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
m01 = pd.read_csv("meter_01_all.csv",parse_dates=['date_time'],delimiter=',') #temperature
m04 = pd.read_csv("meter_04_all.csv",parse_dates=['date_time'],delimiter=',') #consumption
m07 = pd.read_csv("meter_07_all.csv",parse_dates=['date_time'],delimiter=',')
m10 = pd.read_csv("meter_10_all.csv",parse_dates=['date_time'],delimiter=',')
m13 = pd.read_csv("meter_13_all.csv",parse_dates=['date_time'],delimiter=',')
m16 = pd.read_csv("meter_16_all.csv",parse_dates=['date_time'],delimiter=',')
m04.head()
```

```python
m01p = m01.set_index('date_time')
m04p = m04.set_index('date_time')
m07p = m07.set_index('date_time')
m10p = m10.set_index('date_time')
m13p = m13.set_index('date_time')
m16p = m16.set_index('date_time')                     
m04p.head()
```

```python
m01d = m01p.resample('W',how='mean')
m04d = m04p.resample('W',how='sum')
m07d = m07p.resample('D',how='sum')
m10d = m10p.resample('D',how='sum')
m13d = m13p.resample('D',how='sum')
m16d = m16p.resample('D',how='sum')
m04d.head()
```

```python
fig,ax = plt.subplots(figsize=(12,3))
ax.plot(m01d.index, m01d, c='blue')
ax.set_ylabel('TEMP')
ax.set_xlabel('date_time')
#plt.xlim("2014-01-01","2015-01-01")
plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(12,3))
ax.plot(m04d.index, m04d, c='red')
ax.set_ylabel('LOAD')
ax.set_xlabel('date_time')
#plt.xlim("2014-01-01","2015-01-01")
plt.tight_layout()
plt.show()
```

```python
plt.scatter(m01p.resample('W',how='mean'),m04p.resample('W',how='sum'),marker='o')
plt.show()
```
